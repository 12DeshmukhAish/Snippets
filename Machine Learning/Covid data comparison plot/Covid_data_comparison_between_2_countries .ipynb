```python
import pandas as pd

data=pd.read_csv('owid-covid-data.csv')
```

```python

data_india=data[data['location']=='India']
data_india=data_india.reset_index()

data_usa=data[data['location']=='United States']
data_usa=data_usa.reset_index()
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt

import seaborn as sns

plt.figure()

#plt.bar(left=data_india.index,height=data_india['total_cases_per_million'])
plt.bar(left=data_usa.index[78:112],height=(data_usa['total_cases'][78:112])/10,label="(US_total cases)/10 by date")
plt.bar(left=data_india.index[78:112],height=data_india['total_cases'][78:112],label="India_total cases by date")

#plt.grid()

plt.title("Covid-19 Total Cases between 19-03-2020 and 21-04-2020 \n USA vs India")
plt.plot(data_india['total_deaths'][78:112],c='r',alpha=0.75,label='(deaths in India)')
plt.plot(data_usa['total_deaths'][78:112],c='k',alpha=0.75,label='(deaths in US)')
plt.xticks(range(76,113,3),data_usa['date'][79:113:3],rotation=30)

#ax1.tick_params(axis='y',labelcolor='blue',)

#ax1.yticks(range(0.0,1.0),

plt.legend()
plt.show()
```
